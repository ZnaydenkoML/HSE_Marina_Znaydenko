```python
""" 
Описание задания:

Реализуйте структуру данных «Участники спора». Она должна представлять собой список (list) словарей (dict).  

Каждый словарь хранит информацию об отдельном участнике. Словарь содержит пары «ключ-значение» с характеристиками участника: 
- name: наименование - status: статус в процессе (Истец, Ответчик, Третье лицо и т.п.) - inn: ИНН.

Заполнение данных должно быть произведено пользователем через консоль для трех различных участников.  

Пример структуры: 
[  
{"name": 'ООО "Рога и Копыта"', "status": "Истец", "inn": "4545454545"},
{"name": 'Баширов А.А.', "status": "Ответчик", "inn": "32323232323232"},
{"name": 'Петров А.А.', "status": "Третье лицо", "inn": "12121212121212"} 
]  

Выведите полученные данные в консоль. 

"""

# Создадим пустой список, в который будем добавлять участников
participants = []  

print('Заполнение данных об участниках спора') 
print('Всего нужно ввести данные для трeх участников.\n')  

# Цикл на 3 участника 
for i in range(1, 4):  
    print(f'Участник №{i}')   
    
# Запросим наименование  
    name = input('Введите наименование участника (например, ООО \"Рога и Копыта\" или ФИО): ')

# Запросим статус
    status = input('Введите статус участника (например, Истец, Ответчик, Третье лицо): ')   

# Запросим ИНН   
    inn = input('Введите ИНН участника: ')   

# Проверим, что в ИНН только числовые значения
    if not inn.isdigit():  
        print('Внимание: ИНН содержит нечисловые значения. Он будет сохранен как есть.\n')
    else:  print('ИИН принят.\n')   

# Сформируем словарь для текущего участника  
    participant = {  
        'name': name,
        'status': status,
        'inn': inn  }   

# Добавим словарь в общий список  
    participants.append(participant)  

# После цикла выводим итоговую структуру 
    print('\nПолученная структура данных «Участники спора»:\n') 
    print(participants)  
    print('\nПодробный вывод по каждому участнику:') 
    for idx, p in enumerate(participants, start=1):  
        print(f'\nУчастник №{idx}:')  
        print(' Наименование:', p['name'])  
        print(' Статус:', p['status'])  
        print(' ИНН:', p['inn'])
```
