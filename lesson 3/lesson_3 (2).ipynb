```python
'''
ОПИСАНИЕ ЗАДАНИЯ:
Необходимо создать ряд функций для проведения математических вычислений:
- функцию вычисления факториала числа (произведение натуральных числе от 1 до n), 
которая принимает в качестве аргумента число, возвращает его факториал;
- функцию поиска наибольшего числа из трех, которая принимает в качестве агрумента кортеж из трех чисел, возвращает наибольшее из них;
- функцию расчета площади прямоугольного треугольника, которая принимает в качестве аргумента размер двух катетов треугольника и 
возвращает площадь треугольника.

'''

# 1. Функция factorial(n) вычисляет факториал числа n
# Факториал - это произведение всех натуральных чисел от 1 до n
def factorial(n):

# проверяем, что число не отрицательное
    if n < 0:
       raise ValueError('Факториал определен только для положительных чисел')
       
# Определяем начальное значение результата
    result = 1    
    
# Перемножаем все числа от 1 до n включительно
    for i in range(1, n + 1):
        result *= i
        
# Возвращаем готовый результат
    return result

# Ввод значения числа n пользователем
n = int(input('Введите натуральное число: '))
    
# Вычисление факториала и вывод результата
print(f'Факториал числа', n, 'равен', factorial(n))
```

```python
# 2. Функция для поиcка наибольшего числа принимает кортеж из трех чисел и возвращает наибольшее из них
def max_of_three(numbers):
    
    # Проверяем длину кортежа
    if len(numbers) != 3:
        raise ValueError('Кортеж должен содержать ровно 3 числа')
    
    # Возвращаем наибольшее число из введенных
    return max(numbers)
    
# Просим пользователя ввести три числа
user_input = input('Введите три числа через пробел: ')

# Разобьем введенную строку на части и переведем их формат в числовой
numbers_tuple = tuple(map(int, user_input.split()))

# Проверяем, что введено ровно три числа
if len(numbers_tuple) != 3:
    print('Ошибка: нужно ввести ровно три числа')
else:
    print(f'Введенный кортеж: {numbers_tuple}')
    print(f'Наибольшее число: {max_of_three(numbers_tuple)}')
```

```python
# 3. Функция вычисления площади прямоугольного треугольника
# Функция вычисляет площадь, используя длину двух катетов по формуле:
# S = 1/2*a*b, где a и b значения длины катетов

def triangle_area(cathetus_1, cathetus_2):
    area = 0.5 * cathetus_1 * cathetus_2  
    return area   
    
# Предлагаем пользователю ввести длины катетов 
a = float(input('Введите длину первого катета: ')) 
b = float(input('Введите длину второго катета: '))

# Вычисляем площадь 
result = triangle_area(a, b)  

# Вывод полученного результата 
print(f'Площадь прямоугольного треугольника с катетами {a} и {b} равна {result}')
```

```python
'''
ОПИСАНИЕ ЗАДАНИЯ:
Необходимо создать функцию для генерации текста с адресом суда, которая должна по шаблону генерировать 
шапку для процессуальных документов с реквизитами сторон для отправки.

Пример работы функции:
В арбитражный суд города Москвы
Адрес: 115225, г. Москва, ул. Б. Тульская, 17
Истец: Пупкин Василий Геннадьевич
ИНН 1236182357 ОГРНИП 218431927812733
Адрес: 123534, г. Москва, ул. Водников, 13
Ответчик: ООО “Кооператив Озеро”
ИНН 1231231231 ОГРН 123124129312941
Адрес: 123534, г. Москва, ул. Красивых молдавских партизан, 69
Номер дела А40-123456/2023

Функция должна принимать в качестве аргумента словарь с данными ответчика и
номером дела (ссылка на файл с данными).
● на основании номера дела из списка судов должен быть выбран корректный суд для отправки. 
Данные по арбитражным судам есть в указанном выше файле. Используйте код суда из дела.
● с помощью f-string создайте шаблон для заполнения.
● в качестве истца укажите свои данные.
● в данные по ответчику подставьте данные, переданные в функцию в качестве аргумента.
● в конце шапки подставьте номер дела.

Функция должна возвращать готовую шапку в виде строки.

Также необходимо создать еще одну функцию, которая будет принимать в себя список словарей с данными ответчика. 
Используйте цикл for для генерации всех возможных вариантов этой шапки с вызовом первой функции внутри тела 
цикла for и выводом данных, которые она возвращает в консоль.

'''

# Загрузим данные для выполнения задания
# До загрузки в исходный файл  с данными внесены изменения, а именно удалена строка с командой print("stop")
# Для удобства файл с данными переименован в lesson_2_data_2.py
import sys
sys.path.append(r"C:\Users\Марина\Downloads\lesson_2_data_2.py")
import lesson_2_data_2

from lesson_2_data_2 import courts, respondents

# Осуществим нормализацию данных по судам
# В результате ключами для поиска становятся стандартизированные коды судов, а значениями - словари с названием и адресом судов
def normalize_code(value) -> str:
    return str(value).strip().replace('А', '').zfill(2)

def prepare_courts(courts_raw):
    prepared = {}

    for court in courts_raw:
        if not isinstance(court, dict):
            continue

        code = normalize_code(court.get('court_code', ''))
        name = court.get('court_name')
        address = court.get('court_address')

        if code and name and address:
            prepared[code] = {
                'court_name': name,
                'court_address': address
            }

    return prepared


# Приведем номера дел к единому формату вида A40-180476|2024

def normalize_case_number(case_number):
    if not case_number:
        return None

    case_number = str(case_number).strip().replace(' ', '')

    if case_number.startswith('А'):
        raw = case_number[1:]
    else:
        raw = case_number

    if '-' not in raw:
        return None

    court_code, rest = raw.split('-', 1)
    court_code = normalize_code(court_code)

    if '/' not in rest and len(rest) > 4:
        rest = f'{rest[:-4]}/{rest[-4:]}'

    return f'А{court_code}-{rest}'


def get_case_code(case_number):
    normalized = normalize_case_number(case_number)
    if not normalized:
        return None
    return normalized.split('-')[0].replace('А', '')


# Преобразуем каждую запись из списка respondents в словарь с единообразными ключами: имя, ИНН, ОГРН, адрес и нормализованный номер дела
def prepare_respondent(respondent):
    name = (
        respondent.get('short_name')
        or respondent.get('shortname')
        or respondent.get('full_name')
        or respondent.get('fullname')
        or 'Неизвестный ответчик'
    )

    case_number_raw = respondent.get('case_number') or respondent.get('casenumber')
    case_number = normalize_case_number(case_number_raw)

    return {
        'name': name,
        'inn': respondent.get('inn', 'не указан'),
        'ogrn': respondent.get('ogrn', 'не указан'),
        'address': respondent.get('address', 'не указан'),
        'case_number': case_number
    }


# Создадим справочник судов, с использованием которого по коду суда можно будет получить его название и адрес
prepared_courts = prepare_courts(courts)

# вручную добавим суд, данных о котором не хватило
prepared_courts.setdefault('40', {
    'court_name': 'Арбитражного суда города Москвы',
    'court_address': '115225, г. Москва, ул. Большая Тульская, 17'
})


# С помощью функции генерируем текст с адресом суда по заданному шаблону
PLAINTIFF_NAME = 'Знайденко Марина Леонидовна'
PLAINTIFF_INN = '7700000000'
PLAINTIFF_OGRNIP = '320770000000000'
PLAINTIFF_ADDRESS = '107026, г. Москва, Оружейный пер., д. 41'

def generate_header(respondent: dict) -> str:
    data = prepare_respondent(respondent)

    if not data['case_number']:
        raise ValueError(f'У ответчика {data['name']} отсутствует или некорректен номер дела')

    court_code = get_case_code(data['case_number'])

    if court_code not in prepared_courts:
        raise ValueError(f'Суд с кодом {court_code} не найден')

    court = prepared_courts[court_code]

    header = (
        f'В {court['court_name']}\n'
        f'Адрес: {court['court_address']}\n\n'
        f'Истец: {PLAINTIFF_NAME}\n'
        f'ИНН {PLAINTIFF_INN} ОГРНИП {PLAINTIFF_OGRNIP}\n'
        f'Адрес: {PLAINTIFF_ADDRESS}\n\n'
        f'Ответчик: {data['name']}\n'
        f'ИНН {data['inn']} ОГРН {data['ogrn']}\n'
        f'Адрес: {data['address']}\n\n'
        f'Номер дела {data['case_number']}'
    )

    return header


# C помощью функции пройдемся по всему списку ответчиков, в том числе, если в списке есть "битые" записи
def generate_headers_for_all(respondents_list):
    success = 0
    errors = []

    for respondent in respondents_list:
        try:
            print(generate_header(respondent))
            print('-' * 80)
            success += 1
        except Exception as e:
            errors.append(str(e))

    print(f'Успешно обработано: {success}')
    print(f'Ошибок: {len(errors)}')

    if errors:
        print('\nПервые 10 ошибок:')
        for err in errors[:10]:
            print('-', err)
```
