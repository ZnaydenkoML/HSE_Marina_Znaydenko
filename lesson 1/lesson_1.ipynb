# Задание 2

```python
"""
Описание задания:

Пользователь вводит время в секундах. Рассчитайте время и сохраните отдельно в каждую переменную количество часов, минут и секунд. 
Переведите время в часы, минуты, секунды и сохраните в отдельных переменных.

Используйте приведение типов для перевода строк в числовые типы. Предусмотрите проверку строки на наличие только чиловых данных
через встроенный строковый метод .isdigit().

Выведите рассчитанные часы, минуты и секунды по отдельности в консоль.

"""

# Просим пользователя ввести время в секундах
seconds = input('Введите время в секундах:')

# Проверим, что введены только числовые данные
if seconds.isdigit():
    total_seconds = int(seconds)
    
# Посчитаем часы
    hours = total_seconds // 3600

# Вычислим остаток секунд после вычета часов
    remaining_after_hours = total_seconds % 3600

# Посчитаем минуты
    minutes = remaining_after_hours // 60

# Вычислим остаток секунд после вычета минут
    seconds = remaining_after_hours % 60

# Выведем полученные результаты
    print('Часы:', hours)
    print('Минуты:', minutes)
    print('Секунды:', seconds)
else:
    print('Ошибка: нужно ввести только число секунд, без других символов.') 
    
```

# Задание 3

```python
"""
Описание задания:

Запросите у пользователя через консоль число n (от 1 до 9). Найдите сумму чисел n + nn + nnn.

Например, пользователь ввел число 3. Считаем 3 + 33 + 333 = 369.

Выведите данные в консоль.

"""

# Запрашиваем у пользователя число n
n_str = input('Введите одно число n от 1 до 9:')

# Проверим, что это одно число и, что оно от 1 до 9
if n_str.isdigit() and 1 <= int(n_str) <=9:

    # Формируем строки 'n', 'nn', 'nnn'
    nn_str = n_str * 2
    nnn_str = n_str * 3

    # Преобразуем полученные строки в числа
    n = int(n_str)
    nn = int(nn_str)
    nnn = int(nnn_str)

    # Посчитаем сумму полученных чисел

    res = n + nn + nnn

    # Выведем полученные числовые значения и их сумму
    print('n =', n)
    print('nn =', nn)
    print('nnn =', nnn)
    print('Сумма n + nn + nnn =', res)
else:
    print('Ошибка, нужно ввести одно число от 1 до 9.')
```
